**4.decision treee model training**

splitting the data and training the model

```python
x=data.drop(["Species"],axis=1)
y=data["Species"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=101)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
x_test.shape
```

```python
y_test.shape
```

```python
#decision tree classifier
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
clf=DecisionTreeClassifier()
clf.fit(x_train,y_train)
pred_tree=clf.predict(x_test)
```

**5.predicting the model**

```python
prediction=clf.predict(x_test)
```

```python
from sklearn import metrics
print('The accuracy of the decision tree is', metrics.accuracy_score(prediction,y_test)*100)
```

**6 visualizing decision tree**

```python
# Putting the feature names and class names into variables
feature = ['sepal length (cm)','sepal width (cm)','petal length (cm)','petal width (cm)']
cl=['setosa', 'versicolor', 'virginica']
```

```python
fig,axes =plt.subplots(nrows = 1, ncols = 1, figsize = (3,3), dpi = 300)
tree.plot_tree(clf,feature_names = feature,class_names=cl,filled = True)
plt.show()
```
